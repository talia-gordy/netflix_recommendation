#Getting Feautures

```python
df_features = df[['title','type','description','release_year','age_certification','genres','seasons','imdb_score']].fillna("")
```

```python
df_features
```

```python
df_features = df_features.rename(columns={'release_year': 'release_date', 'age_certification': 'rating'})
display(df_features.head())
```

```python
df_features.shape
```

```python
df_features
```

```python
x = df_features['title'].astype(str) + ' ' + df_features['type'].astype(str) + ' ' + df_features['release_date'].astype(str) + ' ' + df_features['rating'].astype(str) + ' ' + df_features['genres'].astype(str) + ' ' + df_features['seasons'].astype(str) + ' ' + df_features['imdb_score'].astype(str)
```

```python
x
```

```python
x.shape
```

# Conversion to Tokens

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
tfidf = TfidfVectorizer()
```

```python
x = tfidf.fit_transform(x)
```

```python
x.shape
```

```python
print(x)
```

# Get Similarity Score

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
Similarity_Score = cosine_similarity(x)
```

```python
Similarity_Score
```

```python
Similarity_Score.shape
```

# Get Movie/TV show Name Validation from User and Validate for Closest Spelling

```python
Favorite_Movie_TVshow_Name = input("Enter your favorite movie or TV show: ")
```

```python
All_Movies_TVshow_List = df['title'].astype(str).tolist()
```

```python
import difflib
```

```python
Movie_TVshow_Recommendation = difflib.get_close_matches(Favorite_Movie_TVshow_Name, All_Movies_TVshow_List)
print(Movie_TVshow_Recommendation)
```

```python
Close_Match = Movie_TVshow_Recommendation[0]
print(Close_Match)
```

```python
Index_of_Close_Match_Movie_TVshow = df[df.title == Close_Match]['id'].index[0]
print(Index_of_Close_Match)
```

```python
Recommendation_Score = list(enumerate(Similarity_Score[Index_of_Close_Match_Movie_TVshow]))
print(Recommendation_Score)
```

```python
len(Recommendation_Score)
```

# Get all Movies and TV Shows Based on Recommendation Score with Favorite Movie/TV show

```python
Sorted_Similar_Movies_TVshows = sorted(Recommendation_Score, key = lambda x:x[1], reverse = True )
print(Sorted_Similar_Movies_TVshows)
```

```python
print("Top 30 Movies/Tv shows suggested for you: \n")
i = 1
for movie_tvshow_tuple in Sorted_Similar_Movies_TVshows:
  index = movie_tvshow_tuple[0]
  if index == Index_of_Close_Match_Movie_TVshow:
    continue
  title_from_index = df.loc[index, 'title']
  if i <= 30:
    print(f"{i}. {title_from_index}")
    i += 1
  else:
    break
```

```python
Movie_TVshow_Name = input("Enter your Favorite Movie or TV show name: ")
list_of_all_titles = df['title'].astype(str).tolist()

exact_match_found = False
if Movie_TVshow_Name in list_of_all_titles:
    Close_Match = Movie_TVshow_Name
    exact_match_found = True
    print(f"Generating recommendations for: {Close_Match}")
else:
    Find_Close_Match = difflib.get_close_matches(Movie_TVshow_Name, list_of_all_titles)
    if Find_Close_Match:
        Close_Match = Find_Close_Match[0]
        print(f"Exact match for '{Movie_TVshow_Name}' not found. Generating recommendations for the closest match: '{Close_Match}'")
    else:
        print(f"Sorry, '{Movie_TVshow_Name}' and similar titles were not found in the database. Please try another name.")
        Close_Match = None

if Close_Match:
    Index_of_Movie_TVshow = df[df.title == Close_Match].index[0]
    Recommendation_Score = list(enumerate(Similarity_Score[Index_of_Movie_TVshow]))
    sorted_similar_movies_tvshows = sorted(Recommendation_Score, key = lambda x:x[1], reverse = True)
    print("\nTop 10 Movies/TV shows suggested for you: ")
    i = 1
    for movie_TVshows in sorted_similar_movies_tvshows:
      index = movie_TVshows[0]
      if df.index[index] != Index_of_Movie_TVshow:
          title_from_index = df.loc[index, 'title']
          if (i <= 10):
            print(i, '.', title_from_index)
            i += 1
          else:
            break
```
